```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv('sample_clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df[df['ASSESSED_VALUE'] != 0]
df.describe()  # data cleaning. delete the row where ASSESSED_VALUE == 0
```

```python
df_mean = df.groupby('ROLL_NUMBER').mean()
df_mean
```

```python
df_mean.sort_values(by='ASSESSED_VALUE', ascending=False)
```

## General Visualiziton

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.set_title('Box plot of the prices')
sns.boxplot(x='ASSESSED_VALUE', data = df_mean)
```

I found there are some outliers although the data is pretty clean.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Distribution of the prices')
sns.distplot(df_mean['ASSESSED_VALUE'], bins=30, kde=False)
```

```python
df_cheap= df_mean[df_mean['ASSESSED_VALUE'] < 1000000]
df_cheap.describe()
```

```python

df_mean['price_catg'] = df_mean['ASSESSED_VALUE'].apply(lambda x: 'less than 200000' if x <= 200000 else '200000~400000' if (200000<x) & (x<=400000) else '400000~600000' if (400000<x )& (x<=600000) else '800000~1000000' if (800000<x) & (x<=1000000) else '1000000~2000000' if (1000000<x) & (x<=2000000) else 'more than 2000000')
sns.set_palette("Reds")
fig, ax = plt.subplots(figsize=(20,15))
ax.set_title('Geographical distribution of the houses colored by prices')
sns.scatterplot(x= 'LONGITUDE', y='LATITUDE', data = df_mean, hue = 'price_catg', ax=ax )
```

```python

fig, ax = plt.subplots(figsize=(20,15))
ax.set_title('Box plot of the prices on fsa')
sns.boxplot(x='fsa', y='ASSESSED_VALUE', data = df)
```

```python
df_Y = df.groupby('Year')
df_Y.describe()
```

```python
fig, axs = plt.subplots(1, figsize=(14, 10))
df_safCor = pd.concat([df_mean['ASSESSED_VALUE'],df_mean['saf1'],df_mean['saf2'],df_mean['saf3'],df_mean['saf4'],df_mean['saf5'],df_mean['saf6'],df_mean['saf7'],df_mean['saf8'],df_mean['mr5y']], axis=1)
df_safCor.columns = ['ASSESSED_VALUE','saf1','saf2','saf3','saf4','saf5','saf6','saf7','saf8','mr5y']
sns.heatmap(df_safCor.corr(), annot = True, fmt='.1g')
```

```python
fig, axs = plt.subplots(1, figsize=(14, 10))
df4 = pd.concat([df_mean['ASSESSED_VALUE'],df_mean['Inflation'],df_mean['pop1'],df_mean['pop2'],df_mean['saf3'],df_mean['pop4'],df_mean['pop5'],df_mean['pop6'],df_mean['lan1'],df_mean['lan2']], axis=1)
df4.columns = ['ASSESSED_VALUE','Inflation','pop1','pop2','pop3','pop4','pop5','pop6','lan1','lan2']
sns.heatmap(df4.corr(), annot = True, fmt='.1g')
```

```python
fig, axs = plt.subplots(1, figsize=(14, 10))
df5 = pd.concat([df_mean['ASSESSED_VALUE'],df_mean['inc1'],df_mean['inc2'],df_mean['inc3'],df_mean['inc4'],df_mean['own1'],df_mean['own2'],df_mean['own3'],df_mean['own4'],df_mean['lab1'],df_mean['lab2']], axis=1)
df5.columns = ['ASSESSED_VALUE','inc1','inc2','inc3','inc4','own1','own2','own3','own4','lab1','lab2']
sns.heatmap(df5.corr(), annot = True, fmt='.1g')
```
